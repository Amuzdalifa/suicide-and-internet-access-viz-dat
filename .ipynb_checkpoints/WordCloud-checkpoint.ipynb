```python
mask = np.array(Image.open("rope.png"))
```

```python
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
```

```python
# Transform your mask into a new one that will work with the function:
transformed_wine_mask = np.ndarray((mask.shape[0],mask.shape[1]), np.int32)

for i in range(len(mask)):
    transformed_wine_mask[i] = list(map(transform_format, mask[i]))
```

```python
def red_color_func(word, font_size, position, orientation, random_state=None,
                    **kwargs):
    return "hsl(8, 100%%, %d%%)" % random.randint(50, 50)
```

```python
word_occurences['tolol'] = 10
```

```python
def wordcloud(tweets, title):
    stopwords = set(STOPWORDS)
    stopwords.update(manual_stop_words)
    wordcloud = WordCloud(width=512, height=512, mode="RGBA", background_color="rgba(0,0,0,0)", max_font_size=25, min_font_size=4,
                          max_words=200, stopwords=stopwords, scale = 3,
                          random_state=2018, mask=transformed_wine_mask).generate_from_frequencies(word_occurences)
    plt.figure(title, figsize=(20, 10), facecolor='white', edgecolor='blue')
    plt.imshow(wordcloud.recolor(color_func=red_color_func, random_state=3),
           interpolation="bilinear")
    plt.axis("off")
    plt.title(title)
    plt.tight_layout(pad=0)
    wordcloud.to_file("wordcloud-new.png")
    plt.show()
wordcloud(comments, "komentar netizen")
```
